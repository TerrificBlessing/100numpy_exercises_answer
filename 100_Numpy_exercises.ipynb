#### 63. Create an array class that has a name attribute (★★☆)

````python
hint(63)
class NamedArray(np.ndarray):
    def __new__(cls, array, name="no name"):
        obj = np.asarray(array).view(cls)
        obj.name = name
        return obj
    def __array_finalize__(self, obj):
        if obj is None: return
        self.name = getattr(obj, 'name', "no name")

Z = NamedArray(np.arange(10), "range_10")
print (Z.name)
answer(63)
````

#### 78. Consider 2 sets of points P0,P1 describing lines (2d) and a point p, how to compute distance from p to each line i (P0[i],P1[i])? (★★★)

````python
hint(78)
#Reference: https://mathworld.wolfram.com/Point-LineDistance2-Dimensional.html
P0 = np.random.uniform(-10,10,(10,2))
P1 = np.random.uniform(-10,10,(10,2))
p  = np.random.uniform(-10,10,( 1,2))

def distance(P0,P1,p):
    v = P1 - P0
    v[:, [0, 1]] = v[:, [1, 0]]
    v[:, 0] *= -1 # 此时v是线段P0P1的vector perpendicular（法向量）
    norm = np.linalg.norm(v, axis=1) # 法向量的模
    r = p - P0
    d = np.abs(np.sum(v * r, axis=1) / norm)
    return d

print(distance(P0, P1, p))

# answer:
def distance_faster(P0,P1,p):
    #Author: Hemanth Pasupuleti
    #Reference: https://mathworld.wolfram.com/Point-LineDistance2-Dimensional.html

    v = P1- P0 
    v[:,[0,1]] = v[:,[1,0]]
    v[:,1]*=-1 
    norm = np.linalg.norm(v,axis=1)   
    r = P0 - p
    d = np.abs(np.einsum("ij,ij->i",r,v)) / norm 

    return d
print(distance_faster(P0, P1, p))

answer(78)
````

#### 80. Consider an arbitrary array, write a function that extracts a subpart with a fixed shape and centered on a given element (pad with a `fill` value when necessary) (★★★)

````python
hint(80)
Z = np.random.randint(0,10,(10,10))
shape = (5,5)
fill  = 0
position = (1,1)

R = np.ones(shape, dtype=Z.dtype)*fill
P  = np.array(list(position)).astype(int)
Rs = np.array(list(R.shape)).astype(int)
Zs = np.array(list(Z.shape)).astype(int)

R_start = np.zeros((len(shape),)).astype(int)
R_stop  = np.array(list(shape)).astype(int)
Z_start = (P-Rs//2)
Z_stop  = (P+Rs//2)+Rs%2

R_start = (R_start - np.minimum(Z_start,0)).tolist()
Z_start = (np.maximum(Z_start,0)).tolist()
R_stop = np.maximum(R_start, (R_stop - np.maximum(Z_stop-Zs,0))).tolist()
Z_stop = (np.minimum(Z_stop,Zs)).tolist()

r = tuple([slice(start,stop) for start,stop in zip(R_start,R_stop)])
z = tuple([slice(start,stop) for start,stop in zip(Z_start,Z_stop)])
R[r] = Z[z]
print(Z)
print(R)
answer(80)
````

#### 85. Create a 2D array subclass such that Z[i,j] == Z[j,i] (★★★)

````python
hint(85)
class Symetric(np.ndarray):
    def __setitem__(self, index, value):
        i,j = index
        super(Symetric, self).__setitem__((i,j), value)
        super(Symetric, self).__setitem__((j,i), value)

def symetric(Z):
    return np.asarray(Z + Z.T).view(Symetric)

S = symetric(np.random.randint(0,10,(5,5)))
print(S)
S[2,3] = 42
print(S)

answer(85)
````

#### 88. How to implement the Game of Life using numpy arrays? (★★★)

````python
hint(88)
def iterate(Z):
    N = (Z[0:-2,0:-2] + Z[0:-2,1:-1] + Z[0:-2,2:] +
         Z[1:-1,0:-2]                + Z[1:-1,2:] +
         Z[2:  ,0:-2] + Z[2:  ,1:-1] + Z[2:  ,2:]) # N是除了边界的全部细胞的周围存活细胞数
    
    birth = (N==3) & (Z[1:-1,1:-1]==0) # 死细胞是否变活
    survive = ((N==2) | (N==3)) & (Z[1:-1,1:-1]==1) # 活细胞是否继续存活
    Z[...] = 0 # 首先全部清零
    Z[1:-1,1:-1][birth | survive] = 1 # 只考虑活细胞的出现，因为其他情况肯定为0
    return Z

Z = np.random.randint(0, 2, (6, 6))
print(Z)
for i in range(3): 
    Z = iterate(Z)
    print(Z)
answer(88)
````

#### 90. Given an arbitrary number of vectors, build the cartesian product (every combination of every item) (★★★)

````python
hint(90)
def cartesian(arrays):
    arrays = [np.asarray(a) for a in arrays]
    shape = (len(x) for x in arrays)
    ix = np.indices(shape, dtype=int) # 生成索引网格
    ix = ix.reshape(len(arrays), -1).T
    for n, arr in enumerate(arrays):
        ix[:, n] = arrays[n][ix[:, n]]

    return ix

print (cartesian(([1, 2, 3], [4, 5], [6, 7])))
answer(90)
````
